## Create model on numeric and object data types separately
**1. Numeric data type**

**Select only Numeric data types**

```python
num_col = df.select_dtypes(include=[np.number]).columns.to_list()
print(num_col)
```

```python
df_num = df[num_col].copy()
```

## Test Train split

```python
X = df_num.drop(columns=['price'])
y = df_num['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.20,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Model building using pipeline

**Linear Regression**

```python
pipeline = Pipeline([('tranform',PolynomialFeatures(degree = 3,include_bias=False)),
                     ('scale', StandardScaler()),
                     ('regres', LinearRegression())])
pipeline.fit(X_train, y_train)
train_predict = pipeline.predict(X_train)
test_predict = pipeline.predict(X_test)
mse_train = mean_squared_error(y_train, train_predict)
mse_test = mean_squared_error(y_test, test_predict)
print(mse_train)
print(mse_test)
print(pipeline.named_steps['regres'].coef_)
print(X.columns)
```

**Plot the results**

```python
x = list(range(0, len(y_test)))
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x = x, y = y_test, label = "actual_values");
sns.lineplot(x = x, y = test_predict, label = "predicted_values");

plt.title("Comparison of Real vs Predicted: Linear Regression");
```

**Ridge Regression**

```python
pipe_ridge = Pipeline([('transform', PolynomialFeatures(degree = 3, include_bias=False)),
                       ('scale', StandardScaler()),
                       ('ridge', Ridge())])
pipe_ridge.fit(X_train, y_train)
train_predict_ridge = pipe_ridge.predict(X_train)
test_predict_ridge = pipe_ridge.predict(X_test)
mse_train_ridge = mean_squared_error(y_train, train_predict_ridge)
mse_test_ridge = mean_squared_error(y_test, test_predict_ridge)
print(mse_train_ridge)
print(mse_test_ridge)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x = x, y = y_test, label = "actual_values");
sns.lineplot(x = x, y = test_predict_ridge, label = "predicted_values");

plt.title("Comparison of Real vs Predicted: Ridge Regression");
```

## Grid Search CV

```python
untrained_ridge_model = Pipeline([('transform', PolynomialFeatures(degree = 3, include_bias=False)),
                       ('scale', StandardScaler()),
                       ('ridge', Ridge())])
param_dict = {'ridge__alpha': [0.001, 0.1, 1.0, 10.0, 100.0, 1000.0]}
selector = GridSearchCV(estimator=untrained_ridge_model,
                       scoring = 'neg_mean_squared_error',
                       param_grid=param_dict)
selector.fit(X_train, y_train)
train_predict_grid = selector.predict(X_train)
test_predict_grid = selector.predict(X_test)
mse_train_grid = mean_squared_error(y_train, train_predict_grid)
mse_test_grid = mean_squared_error(y_test, test_predict_grid)
print(mse_train_grid)
print(mse_test_grid)
best_model = selector.best_estimator_
print(best_model)

```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x = x, y = y_test, label = "actual_values");
sns.lineplot(x = x, y = test_predict_grid, label = "predicted_values");

plt.title("Comparison of Real vs Predicted: Grid Search CV");
```

**conclusion: Among all the regression models we tried, the LinearRegression seems to be the best one so far
